```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("C:/Users/Yahoo/OneDrive/Desktop/JUPYTER NOTEBOOK/archive/Heart_Disease (1).csv")
missing_values = data.isnull().sum()

# To check for missing values in a specific column, you can use:
# missing_values = df['column_name'].isnull().sum()

# Print the columns with missing values and their respective counts
print("Missing Values:")
print(missing_values[missing_values > 0])
```

```python
data['sex'].fillna(data['sex'].mode()[0])

# Impute missing values in 'smokingStatus' with mode
data['smokingStatus'].fillna(data['smokingStatus'].mode()[0])
data.isnull().sum()
data.head()
```

```python
import seaborn as sns
ax = sns.countplot(data=data, x="sex", hue="CHDRisk")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from above data we analysed that number of female are more than number of males in the data
#males are more prone to develop a CHDrisk
```

```python
age_bins = [30, 40, 50, 60, 70]
age_labels = ['30-39', '40-49', '50-59', '60-69']
data['age_group'] = pd.cut(data['age'], bins=age_bins, labels=age_labels, right=False)

# Plot the data
ax = sns.countplot(data=data, x="age_group", hue="CHDRisk")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of CHD Risk by Age Group')
plt.show()
```

```python
#CHD risk is high between age group 50-59
```

```python
chol_bins = [150, 200, 250, 300, 350]
chol_labels = ['150-200', '200-250', '250-300', '300-350']
data['chol_range'] = pd.cut(data['totChol'], bins=chol_bins, labels=chol_labels)

# Plotting
ax = sns.countplot(data=data, x="chol_range", hue="CHDRisk")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.xlabel('Cholesterol Range')
plt.ylabel('Count')
plt.title('Cholesterol Range vs CHD Risk')
plt.show()
```

```python
data['CHDRisk_numeric'] = data['CHDRisk'].apply(lambda x: 1 if x == 'yes' else 0)

# Select numeric columns for correlation calculation
numeric_columns = ['sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose', 'CHDRisk_numeric','totChol']
correlation_matrix = data[numeric_columns].corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
gb = data.groupby("CHDRisk").agg({
    "totChol": 'mean',
    "sysBP": 'mean',
    "diaBP": 'mean',
    "BMI": 'mean',
    "heartRate": 'mean',
    "glucose": 'mean',
    "cigsPerDay": 'mean',
    "age":'mean'
    
})
print(gb)
```
